```python
#2,A #linear regression

m,b = np.polyfit(df_new.tot_cases,df_new.tot_death,deg=1)
z = [float(entry)*m+b for entry in df_new.tot_cases]
    
#plotting
plt.plot(df_new.tot_cases,z,color='black', label = 'Total Cases Linear Regression Line')
    
#plot total number of cases vs total number of deaths per state
plt.scatter(df_new.tot_cases, df_new.tot_death, label = 'Total cases vs Total deaths')
    
#labels
plt.xlabel('Total Number of cases')
plt.ylabel('Total number of deaths per state')
plt.title('total number of cases vs total number of deaths per state linear regression')
plt.legend()
plt.show()
```

```python
#polynimal regression
#2b
x = df_new.tot_cases
y = df_new.tot_death
coefficients = np.polyfit(x, y, 3)
poly = np.poly1d(coefficients)
    
new_x = np.linspace(x,y)
new_y = poly(new_x)
    
#plt.figure(2)
plt.plot(x, y, "o", new_x, new_y);

plt.xlabel('Total number of cases')
plt.ylabel('Total number of deaths per state')
plt.title('total number of cases vs total number of deaths per state polynomial regression')
plt.legend(['Total Cases vs Total Death', 'Total Cases vs Total Death polynomial regression'])
plt.show()
```

```python
#4. Consider data for the following states: Washington, Kentucky, Georgia, and California. 
#a) Plot new cases over time. 
#b) Plot total cases over time.  
#c) Find population count for each state and normalize new cases data and total cases data. 
#   Replot a) and b).  
#d) Discuss the results. 


#Washington, Kentucky, Georgia, and California
states = ['WA', 'KY', 'GA', 'CA']
df_state = df;

#sorting
df_state.sort_values(by='submission_date',inplace=True);
df_state = df_state[df_state['state'].isin(states)];

#change df in order
df_state['submission_date'] = pd.to_datetime(df_state['submission_date']);
```

```python
#df_state
#ordered
```

```python
#New Cases over Time 'submission date'
ax = df_state.plot(x='submission_date', 
              y='new_case', 
              title='New Cases over Time');
ax.set_xlabel("Submission Date")
ax.set_ylabel("New Cases")
```

```python
#New Cases over Time 'submission date'
ax = df_state.plot(x='submission_date', 
              y='tot_cases', 
              title='Total Cases over Time');
ax.set_xlabel("Submission Date")
ax.set_ylabel("Total Cases")
```

```python
#New Cases over Time 'submission date'
#normalized data

#Populations
pop = {'WA': 7793612, 'KY': 4504836, 'GA': 10711908, 'CA': 39538223}

#normalize new cases and total cases
for state in states:
    df_state.loc[df_state['state'] == state, 'new_cases_n'] = df_state['new_case'] / pop[state] * 100000;
    df_state.loc[df_state['state'] == state, 'tot_cases_n'] = df_state['tot_cases'] / pop[state] * 100000;
```

```python
ax = df_state.plot(x='submission_date', 
              y='new_cases_n', 
              title='New Cases over Time Normalized per 100,000');
ax.set_ylabel("New Cases Normalized")
ax.set_xlabel("Submission  Date")
```

```python

ax = df_state.plot(x='submission_date', 
              y='tot_cases_n', 
              title='Total cases over Time Normalized per 100,000');
ax.set_xlabel("Population")
ax.set_ylabel("Submission Dates")
```

```python
#problem 3
#A)
a = 3.2
b = 0.6
c = 50
d = 0.56
k = 125
r = 1.6

x1 = 25
y1 = 20

x2 = 1
y2 = 1

x3 = 20
y3 = 2

def fun(t,z,r,a,b,c,d,k):
        x,y = z
        xdot = (r*x)*(1- x/k) - (a*x*y)/(c+x)
        ydot = (b*a*x*y)/(c+x) - (d*y)
        return  xdot, ydot

#print(xdot)
#print(ydot)

s = solve_ivp(fun,[0,50],[x1,y1], args=(r, a, b, c, d, k))
plt.plot(s.t,s.y[0], label='Pop 1')
plt.plot(s.t,s.y[1], label='Pop 2')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Dynamical Population Model (x = 25,  y = 20)')
plt.legend()
plt.show()
```
